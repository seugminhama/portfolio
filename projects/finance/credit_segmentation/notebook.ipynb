##데이터 전처리

```python
#object 타입 라벨링
from sklearn.preprocessing import LabelEncoder

label_encoders = {}

for col in X.select_dtypes(include='object').columns:
  le = LabelEncoder()
  X[col] = X[col].fillna("Unknown")
  X[col] = le.fit_transform(X[col].astype(str))
  label_encoders[col] = le
```

```python
# Y변수 라벨링
le_y = LabelEncoder()
Y_encoded = le_y.fit_transform(Y)
```

```python
from xgboost import XGBClassifier

#변수 중요도 확인을 위한 모델 학습
x_model = XGBClassifier(random_state=42)
x_model.fit(X, Y_encoded)
```

```python
# 변수 중요도 추출
importances = pd.Series(x_model.feature_importances_, index=X.columns)
```

```python
# 상위 100개 변수 추출
top_100_features = importances.sort_values(ascending=False).head(100)
print("📌 상위 100개 변수:")
print(top_100_features)

# 원본 데이터에서 상위 100개 변수만 선택
x_train_top100 = X[top_100_features.index]
x_test_top100 = X[top_100_features.index]
```

```python
# 보기 편하게 DataFrame으로 변환 후 정렬된 순서대로 출력
top_100_df = top_100_features.to_frame(name='Importance')
top_100_df.index.name = 'Feature'

# 상위 100개 변수 전체를 보기 좋게 출력
from IPython.display import display
display(top_100_df)
```

```python
# 컬럼명 리스트 예시로 사용
columns = X.columns

# 다중 키워드 그룹 (모두 포함되어야 함)
must_include_pairs = [('이용', '금액')]

# 단일 키워드 그룹
single_keywords = ['할부', '신용', '이자']

# 최종 조건 필터링
filtered_cols = []
for col in columns:
    # '이용' + '금액' 같이 모두 포함된 경우
    pair_match = any(k1 in col and k2 in col for k1, k2 in must_include_pairs)

    # 단일 키워드만 포함된 경우
    single_match = any(k in col for k in single_keywords)

    # 제외 키워드
    exclude_match = 'R3M' in col or 'R6M' in col

    if (pair_match or single_match) and not exclude_match:
        filtered_cols.append(col)
```

```python
X = X[filtered_cols]
X
```

```python
#결측치 개수 다시 확인
nan = X.isnull().sum() # isnull()은 결측값이면 True, 아니면 False인 데이터프레임을 반환하는 것이고 .sum()을 쓰게 되면 변수별로 True의 개수를 세므로 결론적으로 nan은 Series 형태가 됨.
nan_cols = nan[nan > 0]
nan_cols # 마찬가지로 Series 형태
```

```python
#이상치 처리
#수치형 변수
numeric_cols = X.select_dtypes(include=[np.number]).columns

#각 수치형 변수의 왜도
skewness = X[numeric_cols].skew().sort_values(ascending=False)
print(skewness)
```

```python
#왜도 절댓값 2 이상이면 극단으로 분포
extreme_skew_cols = skewness[skewness.abs() > 2].index.tolist()
evenly_distributed_cols = skewness[skewness.abs() <= 2].index.tolist()

print("극단:", extreme_skew_cols)
print("고르게:", evenly_distributed_cols)
```

```python
#극단의 이상치 처리
#로그 변환
for col in extreme_skew_cols:
  if (X[col] > 0).all():
    X[col] = np.log1p(X[col])
  else:
    upper = X[col].quantile(0.99)
    lower = X[col].quantile(0.01)
    X[col] = X[col].clip(lower, upper)
```

```python
#고른 분포의 변수 이상치 처리
from scipy.stats import zscore

z_scores = np.abs(zscore(X[evenly_distributed_cols]))
outlier_mask = (z_scores > 3)

# 이상치 개수 확인
print("고른 분포 변수 이상치 개수:", outlier_mask.sum().sum())

# 이상치 평균 대체
for i, col in enumerate(evenly_distributed_cols):
    outlier_idx = outlier_mask[:, i]
    X.loc[outlier_idx, col] = X[col].mean()
```

```python
# 수치형 변수 선택
numeric_cols = X.select_dtypes(include=[np.number]).columns

# z-score 계산
z_scores = np.abs(zscore(X[numeric_cols]))

# 변수별 이상치 비율 계산 → Pandas Series로 변환
outlier_ratio = pd.Series((z_scores > 3).sum(axis=0) / len(X), index=numeric_cols)

# 정렬해서 출력
print(outlier_ratio.sort_values(ascending=False))
```

```python
# 동일한 값 가지는 변수 확인
constant_cols = X.columns[X.nunique() <= 1].tolist()

# 중복된 변수 확인
duplicated_cols = X.T[X.T.duplicated()].index.tolist()

print("모든 값이 동일한 열:", constant_cols)
print("중복된 열:", duplicated_cols)
```

```python
cols_to_drop = constant_cols + duplicated_cols
X_train = X.drop(columns=cols_to_drop)
```

```python
# 중요도 0인 변수 목록
zero_importance_cols = important_features[important_features == 0].index.tolist()
print(f"📌 중요도 0인 변수 개수: {len(zero_importance_cols)}")
```

```python
# 학습 데이터에서 제거
x_train_reduced = X_train.drop(columns=zero_importance_cols)

# 결과 확인
print(f"제거 후 학습 데이터 shape: {x_train_reduced.shape}")
```

```python
# 클래스 불균형 해결하기
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42, k_neighbors=1)

X_resampled, Y_resampled = smote.fit_resample(x_train_reduced, Y_encoded)

print("SMOTE 적용:", np.bincount(Y_resampled))
```

##모델링

```python
test = pd.read_csv(test_path)
X_test = test.drop(columns=['ID', 'Unnamed: 0.1'])
```

```python
# 1. object 타입 라벨 인코딩 (test에 train에서 학습한 인코더 사용)
for col in test.select_dtypes(include='object').columns:
    test[col] = test[col].fillna("Unknown")
    if col in label_encoders:
        le = label_encoders[col]
        test[col] = le.transform(test[col].astype(str))
    else:
        print(f"[경고] {col}은 학습에 없던 컬럼이거나 인코더가 없음.")

# 2. 변수 중요도 상위 100개 기준 적용
test = test[top_100_features.index.intersection(test.columns)]

# 3. 키워드 기반 컬럼 필터 적용 (filtered_cols 사용)
test = test[[col for col in filtered_cols if col in test.columns]]

# 4. 특정 컬럼 결측값 0으로 처리 (필요 시)
for col in ['최종유효년월_신용_이용', '최종유효년월_신용_이용가능']:
    if col in test.columns:
        test[col] = test[col].fillna(0)

# 5. 결측값 평균 대체
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
test_imputed = pd.DataFrame(imputer.fit_transform(test), columns=test.columns)

# 6. 이상치 처리 (clip or log1p)
# 왜도 기준 분리
from scipy.stats import zscore
skewness_test = test_imputed.skew().sort_values(ascending=False)
extreme_cols = skewness_test[skewness_test.abs() > 2].index.tolist()
even_cols = skewness_test[skewness_test.abs() <= 2].index.tolist()

# 극단 분포 처리
for col in extreme_cols:
    if (test_imputed[col] > 0).all():
        test_imputed[col] = np.log1p(test_imputed[col])
    else:
        upper = test_imputed[col].quantile(0.99)
        lower = test_imputed[col].quantile(0.01)
        test_imputed[col] = test_imputed[col].clip(lower, upper)

# 고른 분포 이상치 평균 대체
z_scores_test = np.abs(zscore(test_imputed[even_cols]))
for i, col in enumerate(even_cols):
    outliers = z_scores_test[:, i] > 3
    test_imputed.loc[outliers, col] = test_imputed[col].mean()

# 7. 중요도 0인 변수 제거
test_final = test_imputed.drop(columns=zero_importance_cols, errors='ignore')

# 최종 결과 확인
print("✅ Test 전처리 완료. 최종 shape:", test_final.shape)
```
